```python
import matplotlib.pyplot as plt
x1 = [1,2,3]
y1 = [2,4,1]

plt.plot(x1, y1, label = "line 1")

x2 = [1,2,3]
y2 = [4,1,3]

plt.plot(x2, y2, label = "line 2")

plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.title('Two lines on same graph!')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as p

year = [2001,2002,2003,2004,2005]

Ritvik = [90,92,94,95,97]
Angad = [89,91,93,95,98] 

p.plot(year,Ritvik, color='b', linestyle=':', linewidth=5) 

p.plot(year, Angad,color='orange') 

p.xlabel('Year') 
p.ylabel('Marks') 

p.title('Ritvik v/s Angad') 

p.legend(('Ritvik','Angad'),loc='upper right',frameon=True) 
p.show()
```

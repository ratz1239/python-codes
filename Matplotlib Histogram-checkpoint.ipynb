```python
import matplotlib.pyplot as p

ages = [2,5,70,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,20,40]
 
ranges = (0, 100)

bins = 10
  
p.hist(ages, bins, ranges, color = 'green', histtype = 'bar', rwidth = 0.8)

p.xlabel('age')
p.ylabel('No. of people')
p.title('My histogram') 

p.show()
```

```python
import matplotlib.pyplot as p
import numpy as np

label = ['Anil', 'Vikas', 'Dharma', 'Mahen', 'Manish', 'Rajesh']

per = [94,85,45,25,50,54]

index = np.arange(len(label))

p.bar(index, per)

p.xlabel('Student Name', fontsize=20) 
p.ylabel('Percentage', fontsize=20) 
p.xticks(index, label, fontsize=20, rotation=30)

#p.xticks(index, label, fontsize=20, rotation=30)
p.title('Percentage of Marks')

p.show()
```

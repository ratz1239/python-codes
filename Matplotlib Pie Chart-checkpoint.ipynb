```python
import matplotlib.pyplot as plt

activities = ['eat', 'sleep', 'work', 'play']

slices = [3, 7, 8, 6]

colors = ['r', 'y', 'g', 'b']
  

plt.pie(slices, labels = activities, colors=colors,explode = (0, 0, 1, 0), autopct = '%1.1f%%')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'Candidate1', 'Candidate2', 'Candidate3', 'Candidate4' 

votes = [315, 130, 245, 210]

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] 

explode = (1, 0, 0, 0) 

plt.pie(votes, explode=explode, labels=labels, colors=colors, autopct='%0.2f%%', shadow=True, startangle=140)

plt.show()
```
